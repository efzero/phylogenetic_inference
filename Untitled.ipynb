double compute_median(const double* values , unsigned n){
  double x[n] = *values;
  
  vector<int> v(std::begin(x), std::end(x));
  vector<double> v2(std::begin(x), std::end(x));


  if (n % 2 == 0){
    auto m1 = v.begin() + v.size()/2;
    nth_element(v.begin(), m1, v.end());
    auto m2 = v2.begin() + v2.size()/2 -1;
    nth_element(v2.begin(), m2, v2.end());

    return (v[v.size()/2] + v2[v2.size()/2-1])
  }

  else{
    auto m = v.begin() + v.size()/2;
    nth_element(v.begin(), m, v.end());
    return v[v.size()/2];
    
  }

  
}

template <typename Container , typename Predicate >
void erase_if(Container& x, Predicate pred){

  for (U::iterator it = x.begin(); it != x.end(); ++it){
    if (pred(*it)){
      erase(it);
    }
  }
  
}
