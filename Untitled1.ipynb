```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [2.25, 4.5, 9, 13.5, 18]
res_lasso = [31.27, 32.81, 33.94, 34.18, 34.12]
lasso_piccs = [30.16, 31.55, 33.08, 33.66, 33.80]
res_nerp = [35.41, 35.97, 35.89, 35.80, 35.68]
res_nerp_alpha67 = [35.45, 35.85, 35.88, 35.69, 35.54]
PICSS_alpha67 = [30.96, 33.25, 35.61, 36.04, 35.94]
PICSS = [30.17, 31.72, 33.67, 34.44, 34.62]
```

```python
plt.figure(figsize = (6,4))
plt.plot(x, res_lasso, label = 'residual learning-lasso alpha=0.33')
plt.plot(x, lasso_piccs, label = 'PICCS-lasso alpha=0.33')
plt.plot(x, PICSS, label = 'PICCS-TV alpha=0.33')
plt.plot(x, PICSS_alpha67, label = 'PICCS-TV alpha=0.67')
plt.plot(x, res_nerp, label = 'Residual-NeRP alpha=0.33')
plt.plot(x, res_nerp_alpha67, label = 'Residual-NeRP alpha=0.67')
plt.xlabel("regularization strength (lambda)")
plt.ylabel("PSNR")
plt.title("Performance vs Regularization for different Methods")
plt.legend()
```
